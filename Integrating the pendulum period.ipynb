```python
#Integrates the numerical factor to find the period of a pendulum
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
tauIntg = lambda phi0, phi: 1 / np.sqrt(np.sin(phi0 / 2) ** 2 - np.sin(phi / 2) ** 2)
```

```python
phi0Max = 2
result = np.zeros(300)
phi0_values = np.linspace(0.05, 3, 300)
```

```python
for i in range(len(phi0_values)):
    phi0 = phi0_values[i]
    result[i] = np.pi * np.trapz([tauIntg(phi0Max, phi) for phi in np.linspace(0, phi0, 100)], np.linspace(0, phi0, 100))
```

```python
plt.plot(phi0_values, result)
plt.xlabel('phi0')
plt.ylabel('Period')
plt.show()
```
